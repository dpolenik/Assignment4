```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np

vISteps =pd.read_csv('./output-values/small/value-iteration-steps.csv')
vISteps=vISteps.drop(vISteps.columns[0], axis=1)
vISteps=vISteps.values.flatten()
```

```python
plt.xticks(np.arange(0, max(vISteps), 5.0))
plt.title("Value Iteration Steps Small World")
plt.plot(vISteps)
plt.show()
```

```python
valIterMs =pd.read_csv('./output-values/small/value-iteration-milliseconds.csv')
valIterMs=valIterMs.drop(valIterMs.columns[0], axis=1)
valIterMs=valIterMs.values.flatten()
```

```python
plt.title("Value Iteration Time Small World")
plt.plot(valIterMs)
plt.show()
```

```python
valIterRewards =pd.read_csv('./output-values/small/value-iteration-total-reward-gained.csv')
valIterRewards=valIterRewards.drop(valIterRewards.columns[0], axis=1)
valIterRewards=valIterRewards.values.flatten()
```

```python
plt.xticks(np.arange(0, max(vISteps), 5.0))
plt.title("Value Iteration Rewards Small World")
plt.plot(valIterRewards)
plt.show()
```

```python
def mapValues(inputCsv,title):
    valIterRewards =pd.read_csv(inputCsv)
    valIterRewards=valIterRewards.drop(valIterRewards.columns[0], axis=1)
    valIterRewards=valIterRewards.values.flatten()
    plt.xticks(np.arange(0, max(vISteps), 5.0))
    plt.title(title)
    plt.plot(valIterRewards)
    plt.show()
mapValues('./output-values/small/policy-iteration-total-steps.csv',"Policy Iteration Steps Small World ")
mapValues('./output-values/small/policy-iteration-total-reward-gained.csv',"Policy Iteration Reward Maximums Small World")
mapValues('./output-values/small/policy-iteration-milliseconds.csv',"Policy Iteration Time To Optimal Policy Small World")
```

```python
mapValues('./output-values/small/q-learning-steps.csv',"Q Learning Steps Small World")
mapValues('./output-values/small/q-learning-total-reward-gained.csv',"Q Learning Reward Maximums Small World")
mapValues('./output-values/small/q-learning-milliseconds.csv',"Q Learning Time To Optimal Policy Small World")
```
